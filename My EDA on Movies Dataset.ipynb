```python
filtered_genres = df.set_index('Title').Genre.str.split(',', expand=True).stack().reset_index(level=1, drop=True)
plt.figure(figsize=(10,10))
g = sns.countplot(y = filtered_genres, order=filtered_genres.value_counts().index[:])
plt.title('Genre wise details of the Movies')
plt.xlabel('No of Movies')
plt.ylabel('Genres')
plt.show()
```

```python
df_movie_genre = df
df_movie_genre['genres'] = df['Genre'].apply(lambda x: x.split(",")[0])
df_movie_genre.shape
```

```python
plt.rcParams.update({'font.size': 22})
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(18,15))
sns.set_palette("Set1", 20, .65)
df_movie_genre.groupby(['Year', 'genres']).count()['Title'].unstack().plot.area(ax=ax);
ax.set(xlabel='Year', ylabel='Number of Movies', title = 'Trend of Movie Releases over Time by Genres')
```

```python
genre_year = df_movie_genre.groupby(['genres', 'Year']).mean().sort_index()
genre_year.head(5)
```

```python
df_Revenue = genre_year.pivot_table(index=['genres'], columns=['Year'], values='Revenue', aggfunc=np.mean)
df_Ratings = genre_year.pivot_table(index=['genres'], columns=['Year'], values='Rating', aggfunc=np.mean)
df_Votes = genre_year.pivot_table(index=['genres'], columns=['Year'], values='Votes', aggfunc=np.mean)
f, [axA, axB, axC] = plt.subplots(figsize=(40, 30), nrows=3)
cmap = sns.cubehelix_palette(start=1.3, rot=1.3, as_cmap=True)
sns.heatmap(df_Revenue, xticklabels=5, cmap=cmap, linewidths=0.05, ax=axA)
sns.heatmap(df_Ratings, xticklabels=5, cmap=cmap, linewidths=0.05, ax=axB)
sns.heatmap(df_Votes, xticklabels=5, cmap=cmap, linewidths=0.05, ax=axC)
axA.set_title('Average Revenue over Time for each Genre')
axB.set_title('Average Ratings over Time for each Genre')
axC.set_title('Average Votes Score over Time for each Genre')
axA.set_xlabel('Year')
axA.set_ylabel('Genres')
axB.set_xlabel('Year')
axB.set_ylabel('Genres')
axC.set_xlabel('Year')
axC.set_ylabel('Genres')
plt.show()
```

```python
df_Runtime = genre_year.pivot_table(index=['genres'], columns=['Year'], values='Runtime', aggfunc=np.mean)
df_Metascore = genre_year.pivot_table(index=['genres'], columns=['Year'], values='Metascore', aggfunc=np.mean)
f, [axA, axB] = plt.subplots(figsize=(40, 30), nrows=2)
cmap = sns.cubehelix_palette(start=1.3, rot=1.3, as_cmap=True)
sns.heatmap(df_Revenue, xticklabels=5, cmap=cmap, linewidths=0.05, ax=axA)
sns.heatmap(df_Ratings, xticklabels=5, cmap=cmap, linewidths=0.05, ax=axB)
axA.set_title('Average Runtime over Time for each Genre')
axB.set_title('Average Metascore over Time for each Genre')
axA.set_xlabel('Year')
axA.set_ylabel('Genres')
axB.set_xlabel('Year')
axB.set_ylabel('Genres')
plt.show()
```
